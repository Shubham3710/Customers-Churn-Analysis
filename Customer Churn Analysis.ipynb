```python
# Calculate the count for each combination of 'SeniorCitizen' and 'Churn'
count_data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack().fillna(0)

# Calculate the total count for each 'SeniorCitizen' category
total = count_data.sum(axis=1)

# Calculate the percentage for each 'Churn' category by dividing by the total and multiplying by 100
percentage_data = count_data.div(total, axis=0) * 100

# Plot the stacked bar chart
plt.figure(figsize=(4,4))
ax = percentage_data.plot(kind='bar', stacked=True, color=['#FF6347', '#4682B4'], width=0.8)

# Add percentage labels on the bars
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{height:.1f}%', (p.get_x() + p.get_width() / 2., p.get_height() / 2.),
                    ha='center', va='center', fontsize=12, color='white')

# Set the title and labels
plt.title("Churn by SeniorCitizen (%) (stack bar chart)")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage (%)")
plt.xticks([0, 1], ["No", "Yes"], rotation=0)
plt.legend(title="Churn", bbox_to_anchor = (0.9,0.9))

plt.show()
```

```python
# Assuming you have a DataFrame 'df' with the required columns
columns = ['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 
           'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create a figure with multiple subplots
fig, axes = plt.subplots(2, 4, figsize=(20, 10))  # 2 rows, 4 columns of subplots
axes = axes.flatten()  # Flatten the 2x4 array to make indexing easier

# Loop through the columns and create a countplot for each
for i, column in enumerate(columns):
    sns.countplot(x=column, data=df, ax=axes[i], hue = df["Churn"] )
    axes[i].set_title(f'Countplot of {column}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Adjust layout to prevent overlap
plt.tight_layout()

# Explicitly show the plots in Jupyter
plt.show()
```

#The majority of customers who do not churn tend to have services like PhoneServices,InternetService(partially DSL) and OnlineSecurity enabled.
OnlineBackup, TechSupport and StreamingTV. churn rates are notically higher when these services are not used or are unavailable
